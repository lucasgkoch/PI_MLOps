<h1>user_items<h1\>

Se carga el dataset disponible y se realiza una primera visualizacion

```python
#Se carga el archivo json
filas = list()
with open("raw_data/australian_users_items.json", "r", encoding="Latin-1") as archivo:
    for linea in archivo.readlines():
        filas.append(ast.literal_eval(linea))

df_items = pd.DataFrame(filas)
df_items
```

Se observa que dentro de la columna items hay diccionarios con informacion relevante. decidimos descomprimir esa informacion para tenerla disponible en nuevas columnas

```python
#Se extraen los datos de la columna items en filas
df_items = df_items.explode("items").reset_index()
df_items = df_items.drop(columns="index")

#Se normalizan esos datos en columnas
df_items = pd.concat([df_items, pd.json_normalize(df_items['items'])], axis=1)
df_items.head()
```

```python
#Se elimina la columna items
df_items.drop(columns='items', inplace=True)
```

```python
#Se exporta una pequeÃ±a porcion del dataset en formato .xlsx para utilizar en un posterior EDA
head_items=df_items.head()
head_items.to_excel('head_items.xlsx')
```

```python
#Se guarda el dataset en formato parquet
df_items.to_parquet('processed_data/items_inicial.parquet')
```

<h1>user_Reviews<h1\>


Se carga el dataset disponible y se realiza una primera visualizacion

```python
#Se carga el archivo json
filas = list()
with open("raw_data/australian_user_reviews.json", "r",encoding="Latin-1") as archivo:
    for linea in archivo.readlines():
        filas.append(ast.literal_eval(linea))

df_reviews = pd.DataFrame(filas)
df_reviews
```

```python
#Se extraen los datos de la columna items en filas
df_reviews = df_reviews.explode("reviews").reset_index()
df_reviews = df_reviews.drop(columns="index")

#Se normalizan esos datos en columnas
df_reviews = pd.concat([df_reviews, pd.json_normalize(df_reviews['reviews'])], axis=1)
df_reviews.head()
```

```python
#Se elimina la columna reviews
df_reviews.drop(columns='reviews',inplace=True)
```

```python
#Se exporta una pequeÃ±a porcion del dataset en formato .xlsx para utilizar en un posterior EDA
head_reviews=df_reviews.head()
head_reviews.to_excel('head_reviews.xlsx')
```

```python
#Se guarda el dataset en formato parquet
df_reviews.to_parquet('processed_data/reviews_inicial.parquet')
```
