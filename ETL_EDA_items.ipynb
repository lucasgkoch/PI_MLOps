```python
import pandas as pd
```

Se carga el archivo previamente procesado en ETL_EDA_inicial

```python
#Se carga el archivo parquet
df_items=pd.read_parquet('processed_data/items_inicial.parquet')
df_items.head()
```

Se eliminan registros donde item_counts sea 0, ya que no son utiles para nuestro enfoque

```python
df_items=df_items[df_items['items_count'] != 0]
```

Se revisa la presencia de nulos y duplicados

```python
#Se verifica que no haya valores nulos
df_items.isnull().sum()
```

```python
#Se verifican filas duplicadas en todo el DataFrame
duplicadas = df_items.duplicated()
df_items[duplicadas]
```

```python
#Se eliminan las filas duplicadas
df_items.drop_duplicates(inplace=True)
```

Eliminamos las columnas que no seran necesarias en ninguna parte de nuestro proyecto
<img src="assets/analisis_campos_necesarios_items.png">

```python
#Se eliminan las columnas que no son relevantes para nuestro objetivo
df_items.drop(columns=['items_count','steam_id','user_url','item_name','playtime_2weeks'],inplace=True)
```

```python
#Se guarda el archivo en formato parquet
df_items.to_parquet('processed_data/items.parquet',index=False)
```
